```python
# a.
print("Forma da matriz:", iris.shape)
print("Número de flores:", iris.shape[0])
```

```python
# b.
def to_float_or_nan(x):
    try:
        return float(x)
    except:
        return np.nan

def check_NaNs(matrix):
    print("Existem NaNs!" if np.isnan(matrix).any() else "Não exitem NaNs.")

iris_float = np.vectorize(to_float_or_nan)(iris)
check_NaNs(iris_float)
iris_truncated = iris_float[:, :-1]
print("Forma da Matriz:", iris_truncated.shape)
```

```python
# c.
check_NaNs(iris_truncated)

def print_matrix(matrix):
    columns = ["Comprimento sépala", "Largura sépala", "Comprimento pétala", "Largura pétala"]
    print("+--------------------+----------------+--------------------+----------------+")
    print("| {:} | {:} | {:} | {:} |".format(*columns))
    print("+--------------------+----------------+--------------------+----------------+")
    for row in matrix:
        print("| {:18.2f} | {:14.2f} | {:18.2f} | {:14.2f} |".format(*row))
    print("|                ... |            ... |                ... |            ... |")
    print("+--------------------+----------------+--------------------+----------------+")

print_matrix(iris_truncated[:5])
```

```python
# d.
mean = np.mean(iris_truncated, axis=0)
median = np.median(iris_truncated, axis=0)
std = np.std(iris_truncated, axis=0)

print("Estatísticas por coluna:")
for i, name in enumerate(columns):
    print(f"{name}\n\tMean: {mean[i]:.2f}, Mediana: {median[i]:.2f}, Desvio Padrão: {std[i]:.2f}")
```

```python
# e.
print("iris original:")
print_matrix(iris_truncated[:5])

iris_sorted = iris_truncated[iris[:, 2].argsort()]
print("\niris ordenada:")
print_matrix(iris_sorted[:5])
```

```python
# f.
print("Número total de flores:", iris_truncated.shape[0])
iris_filtered = iris_truncated[(iris_truncated[:, 0] > 1.5) & (iris_truncated[:, 2] < 1.5)]
print("Número de flores filtradas (sépalas > 1.5 & pétalas < 1.5):", iris_filtered.shape[0])
```
