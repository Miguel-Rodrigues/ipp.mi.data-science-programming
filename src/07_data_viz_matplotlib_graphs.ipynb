```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
print('Versão matplotlib:', matplotlib.__version__)
```

## Gráfico de linhas

```python
t = np.arange(0.0, 2.0, 0.01)
v = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, v)
ax.set(xlabel='tempo (s)', ylabel='tensão (mV)', title='Tensão')
ax.grid()

plt.show()
```

## Gráfico de barras

```python
fruits = ['Maça', 'Mirtilo', 'Cereja', 'Laranja']
counts = [40, 100, 30, 55]
bar_labels = ['Vermelha', 'Azul', '_red', 'Laranja']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

fig, ax = plt.subplots()

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
ax.set_ylabel('Abastecimento de fruta')
ax.set_title('Oferta de fruta por tipo e cor')
ax.legend(title='Cor da fruta')

plt.show()
```

## Gráfico de dispersão

```python
np.random.seed(19680801)
x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0
sizes = np.random.rand(*x.shape) * 800 + 500

fig, ax = plt.subplots()
ax.scatter(x, y, sizes, c="green", alpha=0.5, marker=r'$\clubsuit$',
            label="Luck")
ax.set_xlabel("Leprechauns")
ax.set_ylabel("Gold")
ax.legend()
plt.show()
```

## Boxplot

```python
fig, ax = plt.subplots()

np.random.seed(19680801)
all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

ax.boxplot(all_data)
ax.set_title('Boxplot')

ax.yaxis.grid(True)
ax.set_xticks([y + 1 for y in range(len(all_data))],
                labels=['x1', 'x2', 'x3', 'x4'])
ax.set_xlabel('Quatro amostras')
ax.set_ylabel('Valores observados')

plt.show()
```

## Histograma

```python
np.random.seed(19680801)

mu = 100    # média da distribuição
sigma = 15  # dp da distribuição
x = mu + sigma * np.random.randn(437)

num_bins = 50

fig, ax = plt.subplots()

n, bins, patches = ax.hist(x, num_bins, density=True)

# adiciona a linha 'best fit'
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
ax.set_xlabel('Quociente de inteligência')
ax.set_ylabel('Densidade')
ax.set_title(r'Histograma of QI: $\mu=100$, $\sigma=15$')

fig.tight_layout()
plt.show()
```
