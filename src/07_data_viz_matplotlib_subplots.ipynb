````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
````

Dados a utilizar nos gráficos

````python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
````

# O método plt.subplots

Vimos anteriormente que uma forma de criar uma `Figure` e de adicionar `Axes` é através de:
``` python
fig = plt.figure(figsize=(5, 4))
ax = fig.add_axes([1, 1, 1, 1])
```

O método `plt.subplots` cria uma `Figure` e uma grelha de subplots numa única chamada (cada subplot representa um `Axes`), fornecendo controlo razoável acerca de como os gráficos individuais são criados. Para casos de uso mais avançado, deve usar-se `Figure.add_subplot` para adicionar subplots em locais arbitrários dentro da `Figure`.

## Uma figura com um único subplot
A chamada ao método `subplots()` sem argumentos retorna uma `Figure` e um único `Axes`.
Esta é a forma mais simples e recomendada de criar uma única `Figure` e um único `Axes`.

````python
fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title('Um único gráfico')
````

## Criar subplots numa direção
Os dois primeiros argumentos opcionais de `plt.subplots` definem o número de linhas e colunas da grelha.
Ao empilhar apenas numa direção, os `Axes` retornados são uma matriz numpy 1D contendo a lista de `Axes` criados.

````python
fig, ax = plt.subplots(2)
fig.suptitle('Gráficos empilhados na vertical')
ax[0].plot(x, y)
ax[1].plot(x, -y)
````

Se estivermos a criar poucos `Axes`, pode ser  útil descompactá-los imediatamente em variáveis dedicadas para cada `Axes`. Desta forma, podemos usar `ax1`, `ax2`... em vez do `ax[0]`, `ax[1]`, ...

````python
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Gráficos empilhados na vertical')
ax1.plot(x, y)
ax2.plot(x, -y)
````

Para obter gráficos lado a lado devemos usar os dois primeiros parâmetros da `subplots` como ilustrado no exemplo seguinte. Genericamente, pode-se criar uma grelha com o tamanho que se quer especificando os parâmetros `nrows` e `ncols`.

````python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Gráficos lado a lado')
ax1.plot(x, y)
ax2.plot(x, -y)
````

## Criar uma grelha de gráficos
Ao criar uma grelha, os `Axes` retornados são uma matriz NumPy 2D.
Se se tiver que definir parâmetros para cada `subplot`, podemos iterar sobre todos os subplots com um ciclo `for` de:
``` python 
for ax in axs.flat:
    # uso do subplot
```

````python
fig, axs = plt.subplots(nrows=2, ncols=2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')

axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')

axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')

axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Ocultar as labels x e as tick labels nos gráficos superiores
for ax in axs.flat:
    ax.label_outer()
````

Pode-se usar a descompactação de tuplos também em 2D para atribuir todos os gráficos a variáveis dedicadas:

````python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('Grelha 2 x 2')

ax1.plot(x, y)
ax2.plot(x, y**2, 'tab:orange')
ax3.plot(x, -y, 'tab:green')
ax4.plot(x, -y**2, 'tab:red')

for ax in fig.get_axes():
    ax.label_outer()
````
