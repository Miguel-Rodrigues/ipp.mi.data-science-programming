### Exercício 4
Escreva o código Python para gerar o gráfico a seguir reproduzido.<br>
Note que a escala de ambos os eixos se situa entre 1 e 8.<br>

image.png

````python
# Adicione aqui a sua resposta

x = [1, 4, 5, 6, 7]
y = [2, 6, 3, 6, 3]
fig, ax = plt.subplots()
ax.set_xlim(1, 8)
ax.set_ylim(1, 8)
ax.set_xlabel("x - axis")
ax.set_ylabel("y - axis")

ax.plot(x, y, color="red", linestyle = 'dashdot', linewidth=3)
ax.scatter(x, y, color='blue', edgecolors='red', zorder=99, s=200)
ax.set_title('Gráfico com marcadores e linha tracejada')
````

### Exercício 5
Escreva o código Python para gerar um gráfico de barras da popularidade das linguagens de programação.<br><br>
Dados de amostra:<br>
Linguagens de programação: Java, Python, PHP, JavaScript, C#, C++<br>
Popularidade: 22,2, 17,6, 8,8, 8, 7,7, 6,7<br><br>
O código deve gerar um gráfico como o ilustrado de seguida:<br>

image.png

````python
# Adicione aqui a sua resposta

languages = ["Java", "Python", "PHP", "JavaScript",  "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

fig, ax = plt.subplots()
ax.set_xlabel("Linguagens")
ax.set_ylabel("Popularidade")

ax.minorticks_on()
ax.grid(which='major', color='red', linestyle='-')
ax.grid(which='minor', color="black", linestyle=':')
ax.bar(languages, popularity, color="blue", linewidth=10)
ax.set_title('Popularidade de linguagens de programação')
````

### Exercício 10
Escreva o código Python para gerar um gráfico de barras com o número de clientes por estado considerando o dataset público das encomendas efetuadas no site [Olist Store](http://www.olist.com/)<br>
Os dados podem ser descarregados a partir do PAE na pasta `Conteúdos / DADOS / olist_customers_dataset.csv`.<br>

O código deve gerar um gráfico como o ilustrado de seguida:<br>

image.png


````python
# Adicione aqui a sua resposta
from google.colab import files

# Upload ficheiro atravé do colab
uploaded = files.upload()

for fn in uploaded.keys():
  print('Uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

df = pd.read_csv(fn, sep=',')

data = (df[['customer_state', 'customer_unique_id']]
  .groupby('customer_state').count()
  .sort_values('customer_unique_id', ascending=False))

fig, ax = plt.subplots(figsize=(18, 6))
ax.set_xlabel("Estado")
ax.set_ylabel("Número  de Encomendas")
ax.grid(alpha=0.3)
bar_container = ax.bar(data.index, data['customer_unique_id'], color="blue", linewidth=10)
ax.bar_label(bar_container)
ax.set_title('Encomendas por estado')
````
