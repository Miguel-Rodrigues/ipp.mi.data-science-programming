2. Escreva uma função `indicaSalarioSemanal` que recebe o número de horas que o empregado trabalha por semana e o vencimento por hora e que escreve o salário
semanal. No salário semanal, as horas de trabalho acima das 40 são pagas a 1,5 vezes o valor de base (horas extraordinárias). Crie código com input do utilizador para testar a função.

```python
def indicaSalarioSemanal(horasPorSemana: int, vencimentoPorHora: float) -> float:
    if horasPorSemana <= 40:
        return horasPorSemana * vencimentoPorHora
    else:
        return 40 * vencimentoPorHora + (horasPorSemana - 40) * vencimentoPorHora * 1.5

def indicaSalarioSemanal_ai(horasPorSemana: int, vencimentoPorHora: float) -> float:
    return (
        min(horasPorSemana, 40) * vencimentoPorHora +
        max(horasPorSemana - 40, 0) * vencimentoPorHora * 1.5
    )

print("10h/40€: ", indicaSalarioSemanal(10, 40), "€")
print("40h/20€: ", indicaSalarioSemanal(40, 20), "€")
print("60h/10€: ", indicaSalarioSemanal(60, 10), "€")
print("")
print("10h/40€: ", indicaSalarioSemanal_ai(10, 40), "€")
print("40h/20€: ", indicaSalarioSemanal_ai(40, 20), "€")
print("60h/10€: ", indicaSalarioSemanal_ai(60, 10), "€")
```

3. Escrever uma função `tabelaClassificacoes()` que implementa a tabela europeia de classificações para as classificações obtidas em UCs de instituições de ensino superior portuguesas (1-20 valores). A tabela é a seguinte:

| Nota    | Classificação |
| ------- | ------------- |
| 0 - 9   | F             |
| 10 - 12 | E             |
| 13 - 14 | D             |
| 15 - 16 | C             |
| 17 - 18 | B             |
| 19 - 20 | A             |

A função deve receber um inteiro que representa a nota e devolver uma cadeia de carateres que representa a classificação na escala europeia. Se a nota não for válida, a função devolve uma mensagem de erro

```python
def tabelaClassificacoes(nota: int) -> str:
    if 0 <= nota <= 9: return "F"
    if 10 <= nota <= 12: return "E"
    if 13 <= nota <= 14: return "D"
    if 15 <= nota <= 16: return "C"
    if 17 <= nota <= 18: return "B"
    if 19 <= nota <= 20: return "A"
    return "Nota inválida."


for i in range(-1, 22):
    print("Nota", i, ":", tabelaClassificacoes(i))
```

```python
import random

def obtemPositivoInteiro(questao: str, mensagem_validador: str, min_valor: int = 1):
    while True:
        input_value = input(questao)
        if input_value.isdigit():
            value = int(input_value)
            if (value >= min_valor):
                return value

        print(mensagem_validador)

numero_de_alunos = obtemPositivoInteiro(
    "Quantos alunos a turma tem?",
    "Número de alunos tem de ser um valor numérico inteiro maior que 0.")


for i in range(1, numero_de_alunos):
    print("Nota aluno n.", i, ":", tabelaClassificacoes(random.randint(0, 20)))
```

```python
galardoes = [
    ("Física", "Syukuro Manabe, Klaus Hasselmann, Giorgio Parisi"),
    ("Química", "Benjamin List, David W.C. MacMillan"),
    ("Físiologia ou Medicina", "David Julius, Ardem Patapoutian"),
    ("Literatura", "Abdulrazak Gurnah"),
    ("Paz", "Maria Ressa, Dmitry Muratov"),
    ("Ciências Económicas", "David Card, Joshua D. Angrist, Guido W. Imbens")
]

print(galardoes)
```

    <li>Em seguida, escreva uma função que calcula o preço de venda de fruta a um cliente. A função deve receber o inventário da fruta existente para venda na mercearia, a quantidade vendida e imprimir o preço total. O programa deve mostrar uma mensagem de erro caso o produto não exista na mercearia. O programa para quando o utilizador assim o indicar.</li>
</ol>


```python
# a.
inventario = {}

def adicionar_produto(nome: str, preco: float, quantidade: int):
    inventario[nome] = (preco, quantidade)

adicionar_produto("Morango", 5, 20)
adicionar_produto("Laranja", 2.5, 10)
adicionar_produto("Melão", 7.99, 50)

for item, produto in inventario.items():
    print(f"{item}: {produto[0]}€/Kg, Qtd. {produto[1]}")
```

```python
# b.
def calcula_preco(produto: str, quantidade: int):
    if produto in inventario:
        produto_inventario = inventario[produto]
        if produto_inventario[1] > quantidade:
            subtotal = quantidade * produto_inventario[0]
            print(f"{produto}: {produto_inventario[0]}€/Kg, Qtd. {produto_inventario[1]}, subtotal: {subtotal}€")
            return
        else:
            print(f"Produto '{produto}' com quantidade insuficiente para {quantidade}")

    
    print(f"Produto '{produto}' inexistente ou esgotado")

calcula_preco("Morango", 2)
calcula_preco("Laranja", 5)
calcula_preco("Laranja", 30)
calcula_preco("Ananás", 30)
```
