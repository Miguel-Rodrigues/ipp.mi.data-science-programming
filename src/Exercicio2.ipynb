# Ficha Prática 2 - Numpy

##### 1. Use a biblioteca NumPy para criar um vetor de 24 valores aleatórios inteiros entre 10 e 20

```python
import numpy as np

m = np.array(np.random.randint(10, 20, 24))

print("m:", m)
```

###### a. Crie uma cópia do vetor original e substitua os elementos do vetor com índice ímpar com o valor -5 (sem alterar o vetor original).

```python
m2 = m.copy()
m2[1::2] = -5

print("m1:", m)
print("m2:", m2)
```

###### b. Extraia do vetor original os valores entre 10 (inclusive) e 15 (inclusive)

```python
m2 = m.copy()
m2[1::2] = -5

print("m1:", m)
print("m2:", m2)
```
